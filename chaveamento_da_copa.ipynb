```python
# -------------------------------------------
#   ALGORITMO DE CHAVEAMENTO - COPA 2026
#   Feito com listas duplas e dicion√°rio
#   Autor: ChatGPT (vers√£o ilustrativa)
# -------------------------------------------

import random

selecoes = [

    "Canad√°", "M√©xico", "Estados Unidos",
    "Argentina", "Brasil", "Uruguai", "Col√¥mbia", "Equador", "Paraguai",
    "Inglaterra", "Fran√ßa", "Alemanha", "Espanha", "Portugal", "It√°lia",
    "Holanda", "B√©lgica", "Su√≠√ßa", "Cro√°cia", "Dinamarca",
    "Marrocos", "Egito", "Senegal", "Gana", "Nig√©ria", "Costa do Marfim",
    "Jap√£o", "Coreia do Sul", "Ar√°bia Saudita", "Ir√£", "Austr√°lia", "Catar",
    "Uzbequist√£o", "Jord√¢nia",
    "Costa Rica", "Panam√°", "Jamaica",
    "Nova Zel√¢ndia",
    "Vaga_1", "Vaga_2", "Vaga_3", "Vaga_4", "Vaga_5", "Vaga_6", "Vaga_7", "Vaga_8"
]

while len(selecoes) < 48:
    selecoes.append(f"Time_{len(selecoes)+1}")


random.seed(26)
random.shuffle(selecoes)
pots = [selecoes[i*12:(i+1)*12] for i in range(4)]

grupos = {chr(65+i): [] for i in range(12)}


def sortear_grupos(pots, grupos):
    """Distribui as sele√ß√µes dos pots entre os 12 grupos."""
    for pot in pots:
        random.shuffle(pot)
        i = 0
        for selecao in pot:
            grupo_atual = chr(65 + (i % 12))
            grupos[grupo_atual].append(selecao)
            i += 1
    return grupos


resultado = sortear_grupos(pots, grupos)


print("\n=== üèÜ COPA DO MUNDO 2026 - SORTEIO DE GRUPOS ===\n")
for grupo, times in resultado.items():
    print(f"üîπ Grupo {grupo}")
    for time in times:
        print(f"   - {time}")
    print()
```
